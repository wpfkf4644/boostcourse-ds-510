pivot_tableÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ Ïó∞ÎèÑÎ≥Ñ ÌåêÎß§Ïï°Ïùò Ìï©Í≥ÑÎ•º Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî.

Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ Ïó∞ÎèÑÎ≥Ñ ÌåêÎß§Ïï°Ïùò Ìï©Í≥ÑÎ•º Íµ¨Ìï¥ÏÑú Ïñ¥Îäê ÏßÄÏó≠Ïóê ÌåêÎß§Ïï°Ïù¥ ÎßéÏùÄÏßÄÎ•º ÏïåÏïÑÎ≥¥Í≤†ÏäµÎãàÎã§.
pivot_table ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Îã§ÏùåÏùò Í≤∞Í≥ºÎ•º Íµ¨ÌïòÍ≥† ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî!
Í∑∏ÎûòÌîÑÏùò ÏÉâÏÉÅ, Ïä§ÌÉÄÏùºÏùÄ ÏûêÏú†Î°≠Í≤å ÏõêÌïòÎäî Ïä§ÌÉÄÏùºÎ°ú ÏßÄÏ†ïÌï¥ÎèÑ Îê©ÎãàÎã§.

```python
# pivot_tableÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ Ïó∞ÎèÑÎ≥Ñ ÌåêÎß§Ïï°Ïùò Ìï©Í≥Ñ Í≥ÑÏÇ∞
pivot_table = df.pivot_table(
    index=['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ'],
    columns=['Ïó∞ÎèÑ'],
    values='Î∞±Îßå',
    aggfunc='sum'
)

# Í≤∞Í≥º ÌôïÏù∏
pivot_table
```

```python
# Ï∞®Ìä∏ Í∏∏Ïù¥ Ï°∞Ï†à
# Ìï≠ÏÉÅ sns.lineplot Ìò∏Ï∂ú Ï†ÑÏóê Î∞∞Ïπò
plt.figure(figsize=(12, 4))

# Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî 1
# sns.lineplot(data=df_total, x='Ïó∞ÎèÑ', y='Î∞±Îßå', hue='Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ') -> Ïò§Î•ò
# Ïù¥Ïú† : lineplotÏùÄ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÌòïÌÉúÏóêÏÑúÎßå Í∞ÄÎä•. ÌÖåÏù¥Î∏î ÌòïÌÉúÎäî Îç∞Ïù¥ÌÑ∞Î°ú ÏÇ¨Ïö©Ìï† Ïàò ÏóÜÏùå.

# Ïò¨Î∞îÎ•∏ Î∞©Î≤ï
sns.lineplot(data=df, x='Ïó∞ÎèÑ', y='Î∞±Îßå', hue='Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ', errorbar=None, marker ='D')
# Ïù¥Ïú† : dfÎäî Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÌòïÌÉúÏù¥ÎØÄÎ°ú Í∞ÄÎä•

# legend Ìëú Î∞ñÏúºÎ°ú ÎπºÎÇ¥Í∏∞
# sns.lineplot() Ïù¥ÌõÑÏóê Î∞∞Ïπò
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0,)
```

```python
# Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî 2
# ÌîºÎ≤ó ÌÖåÏù¥Î∏î ÌòïÌÉúÎ°ú Ïì∞Î†§Î©¥, ÌåêÎã§Ïä§ Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ÏïºÌïúÎã§.

df_total = df.pivot_table(columns="Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ", index = "Ïó∞ÎèÑ", values = "Î∞±Îßå", aggfunc = 'sum')
df_total.plot(kind='line', figsize=(12, 4), marker='o')

# legend Ìëú Î∞ñÏúºÎ°ú ÎπºÎÇ¥Í∏∞
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0,)
```

```python
# Ïò§Î•òÏΩîÎìú
# df_filtered = df[(df['Ïó∞ÎèÑ'] == 2020) | (df['Ïó∞ÎèÑ'] == 2021) | (df['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ'] == ['Ï§ëÍµ≠','ÎØ∏Íµ≠','ÏùºÎ≥∏'])]

# Ïò¨Î∞îÎ•∏ ÏΩîÎìú
# isinÏùÑ ÏÇ¨Ïö©ÌïúÎã§.
# 2020ÎÖÑÍ≥º 2021ÎÖÑ Îç∞Ïù¥ÌÑ∞Îßå ÌïÑÌÑ∞ÎßÅÌïòÍ≥†, Ï§ëÍµ≠, ÎØ∏Íµ≠, ÏùºÎ≥∏Îßå ÏÑ†ÌÉù
df_filtered = df[(df['Ïó∞ÎèÑ'].isin([2020, 2021, 2022])) & (df['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ'].isin(['Ïó∞ÎèÑ', 'Ï§ëÍµ≠', 'ÎØ∏Íµ≠', 'ÏùºÎ≥∏']))]

# Ï§ëÎ≥µÎêú Îç∞Ïù¥ÌÑ∞Î•º Í∑∏Î£πÌôîÌïòÏó¨ Ìï©Í≥Ñ ÏßëÍ≥Ñ
df_grouped = df_filtered.groupby(['Ïó∞ÎèÑ', 'Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ'], as_index=False)["Î∞±Îßå"].sum()
df_pivot = df_grouped.pivot(index="Ïó∞ÎèÑ", columns="Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ", values="Î∞±Îßå")

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ ÏÉùÏÑ±
result = df_pivot.plot(kind='bar', figsize=(12, 6))

# ÎßâÎåÄ ÏúÑÏóê Í∞í ÌëúÏãú
for container in result.containers:
    result.bar_label(container)

plt.title("2020ÎÖÑ Î∞è 2021ÎÖÑÏùò Ï§ëÍµ≠, ÎØ∏Íµ≠, ÏùºÎ≥∏ Îç∞Ïù¥ÌÑ∞")
plt.xlabel("Ïó∞ÎèÑ")
plt.ylabel("Í∞í")
plt.xticks(rotation=0)  # xÏ∂ï ÎùºÎ≤®ÏùÑ Í∞ÄÎ°úÎ°ú ÌëúÏãú
plt.show()
```

#üìå Q3
groupbyÎ•º ÏÇ¨Ïö©ÌïòÏó¨ 2020ÎÖÑÏùò Ïò®ÎùºÏù∏ Ìï¥Ïô∏ÌåêÎß§ ÏÉÅÌíàÍµ∞ÏùÑ ÏßÄÏó≠Î≥Ñ Ìï©Í≥ÑÎ•º Íµ¨Ìï¥ Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî.

2020ÎÖÑÏùÄ ÏΩîÎ°úÎÇòÎ°ú Ïù∏Ìï¥ Ï†ÑÍµ≠ÎØºÏù¥ Ïñ¥Î†§ÏõÄÏùÑ Í≤™ÏóàÎçò ÏãúÍ∏∞ÏûÖÎãàÎã§. Îòê, ÏïÑÎßàÏ°¥ Îì±Ïùò Ìï¥Ïô∏ Ïò®ÎùºÏù∏ ÌåêÎß§Í∞Ä ÎäòÏñ¥ÎÇú ÏãúÍ∏∞Ïù¥Í∏∞ÎèÑ Ìï©ÎãàÎã§. Ïù¥ Îïå, Ìï¥Ïô∏ Ïñ¥ÎäêÏßÄÏó≠Ïóê Ïò®ÎùºÏù∏ ÌåêÎß§Í∞Ä ÏñºÎßàÎÇò ÎêòÏóàÎäîÏßÄÎ•º groupby() ÏôÄ unstack()ÏùÑ ÌôúÏö©Ìï¥ÏÑú Ìï©Í≥ÑÎ•º Íµ¨ÌïòÍ≥† Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî! Îòê Î∂ÑÏÑùÌïú Í≤∞Í≥ºÎ•º Î≥¥Í≥† Ïñ¥Îñ§ Ïù∏ÏÇ¨Ïù¥Ìä∏Î•º ÏñªÏóàÎäîÏßÄÎèÑ Ìï®Íªò ÏûëÏÑ±Ìï¥ Ï£ºÏÑ∏Ïöî!
Í∑∏ÎûòÌîÑÏùò ÏÉâÏÉÅ, Ïä§ÌÉÄÏùºÏùÄ ÏûêÏú†Î°≠Í≤å ÏõêÌïòÎäî Ïä§ÌÉÄÏùºÎ°ú ÏßÄÏ†ïÌï¥ÎèÑ Îê©ÎãàÎã§.

```python
# 2020ÎÖÑ Îç∞Ïù¥ÌÑ∞Îßå ÌïÑÌÑ∞ÎßÅ
df_2020 = df_filtered[(df_filtered['Ïó∞ÎèÑ'] == 2020) ]

# 'ÏÉÅÌíàÍµ∞Î≥Ñ'Í≥º 'Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ'Î°ú Í∑∏Î£πÌôîÌïòÏó¨ 'Î∞±Îßå' Ïó¥Ïùò Ìï©Í≥ÑÎ•º Í≥ÑÏÇ∞
grouped = df_2020.groupby(['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ', 'ÏÉÅÌíàÍµ∞Î≥Ñ'])['Î∞±Îßå'].sum()

# ÏÉÅÏúÑ 7Í∞ú ÏÉÅÌíàÍµ∞Î≥Ñ ÏÑ†ÌÉù (Ï†ÑÏ≤¥ Ìï©Í≥Ñ Í∏∞Ï§Ä)
top7_groups = grouped.groupby('ÏÉÅÌíàÍµ∞Î≥Ñ').sum().nlargest(7).index

# ÏÉÅÏúÑ 7Í∞ú ÏÉÅÌíàÍµ∞Îßå ÌïÑÌÑ∞ÎßÅ
grouped_top7 = grouped.loc[:, top7_groups]

# Í≤∞Í≥ºÎ•º ÏßÄÏó≠Î≥ÑÎ°ú unstack ÌïòÏó¨ ÎÑìÏùÄ ÌòïÏãùÏúºÎ°ú Î≥ÄÌôò
grouped_unstacked = grouped_top7.unstack()

# Í≤∞Í≥º ÏãúÍ∞ÅÌôî
grouped_unstacked.plot(kind='bar', figsize=(12, 6), colormap='tab10')

# Î≤îÎ°ÄÎ•º Í∑∏ÎûòÌîÑ Î∞ñÏúºÎ°ú ÎπºÍ∏∞
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.title("2020ÎÖÑ ÏßÄÏó≠Î≥Ñ Ïò®ÎùºÏù∏ Ìï¥Ïô∏ÌåêÎß§ ÏÉÅÏúÑ 7Í∞ú ÏÉÅÌíàÍµ∞Î≥Ñ Ìï©Í≥Ñ")
plt.xlabel("Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ")
plt.ylabel("ÌåêÎß§Ïï° (Î∞±Îßå Ïõê)")
plt.xticks(rotation=0)
plt.show()

# unstack, melt ÎπÑÍµê
# unstack: Ïù∏Îç±Ïä§Î•º Ïó¥Î°ú Î≥ÄÌôòÌï©ÎãàÎã§.
# Îç∞Ïù¥ÌÑ∞Î•º ÎÑìÏùÄ ÌòïÌÉúÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
#Ï£ºÎ°ú Îã§Ï§ë Ïù∏Îç±Ïä§Î•º Ï≤òÎ¶¨Ìï† Îïå ÏÇ¨Ïö©Îê©ÎãàÎã§.

# melt:Ïó¥ÏùÑ ÌñâÏúºÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
# Îç∞Ïù¥ÌÑ∞Î•º Í∏∏Í≥† Ï¢ÅÏùÄ ÌòïÌÉúÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
# Ï£ºÎ°ú Îç∞Ïù¥ÌÑ∞Ïùò Ïû¨Íµ¨Ï°∞ÌôîÏóê ÏÇ¨Ïö©Îê©ÎãàÎã§.

# Í≤∞Î°†: unstackÏùÄ Îç∞Ïù¥ÌÑ∞Î•º ÎÑìÍ≤å Î≥ÄÌôòÌïòÍ≥†, meltÎäî Îç∞Ïù¥ÌÑ∞Î•º Í∏∏Í≤å Î≥ÄÌôòÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§.
# Ïù¥ Îëê Ìï®ÏàòÎäî ÏÑúÎ°ú Î∞òÎåÄ Î∞©Ìñ•Ïùò ÏûëÏóÖÏùÑ ÏàòÌñâÌïòÎäî Í≤ΩÏö∞Í∞Ä ÎßéÏúºÎ©∞,
# Îç∞Ïù¥ÌÑ∞Ïùò Íµ¨Ï°∞Î•º Ï°∞Ï†ïÌïòÍ±∞ÎÇò Î≥ÄÌòïÌï† Îïå ÌïÑÏöîÌïú ÏÉÅÌô©Ïóê Îî∞Îùº Ï†ÅÏ†àÌûà ÏÇ¨Ïö©Îê©ÎãàÎã§.
```

```python
# ÎåÄÎ•ôÎ≥Ñ ÌåêÎß§Ïï° Ìï©Í≥Ñ Í≥ÑÏÇ∞
continent_total = df_2020.groupby('Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ')['Î∞±Îßå'].sum()

# Ïõê Í∑∏ÎûòÌîÑ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 8))
plt.pie(continent_total, labels=continent_total.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=140)

plt.title('2020ÎÖÑ ÎåÄÎ•ôÎ≥Ñ Ïò®ÎùºÏù∏ Ìï¥Ïô∏ ÌåêÎß§Ïï° ÎπÑÏú®', fontsize=16)
plt.show()
```

```python
# 60ÏÑ∏ Ïù¥ÏÉÅ ÏÉùÌôúÎπÑ ÏõêÏ≤ú, ÏãúÏ†ê 2020ÎÖÑ
file_path = '/·Ñâ·Ö•·Üº__·Ñã·Öß·Ü´·ÑÖ·Öß·Üº_·ÑÜ·Öµ·Üæ_·Ñâ·Ö¢·Üº·Ñí·Ö™·ÜØ·Ñá·Öµ_·Ñã·ÖØ·Ü´·Ñé·Ö•·Ü´·Ñá·Öß·ÜØ_·ÑÄ·Ö©·ÑÖ·Öß·Üº·Ñå·Ö°_60·Ñâ·Ö¶·Ñã·Öµ·Ñâ·Ö°·Üº_·Ñâ·Öµ·ÑÉ·Ö©_20240825200139.csv'

# CSV ÌååÏùºÏùÑ ÏùΩÏñ¥ÏÑú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
df2 = pd.read_csv(file_path, encoding='cp949')

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò ÏÉÅÏúÑ 5Í∞ú ÌñâÏùÑ Ï∂úÎ†•Ìï©ÎãàÎã§.
df2.head()
```

```python
#ÎèÑÏãúÎ≥Ñ Í≥†Î†πÌôî Ïàò(ÏÑ±Î≥Ñ) Î∂ÑÏÑù

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÎèÑÏãúÎ≥ÑÎ°ú pivot
pivot_df2 = df2.pivot_table(
    index=['ÏÑ±Î≥Ñ', 'Ïó∞Î†πÎ≥Ñ', 'Ìï≠Î™©'],
    values=df2.columns[3:],  # ÎèÑÏãúÎ≥Ñ Ïù∏ÏõêÏàò Ïó¥ ÏÑ†ÌÉù
    aggfunc='sum'  # Ïù∏ÏõêÏàò Ìï©Í≥Ñ Í≥ÑÏÇ∞
).reset_index()

# ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌïú Ï§ÄÎπÑ
pivot_df2_melted = pivot_df2.melt(id_vars=['Ìï≠Î™©', 'ÏÑ±Î≥Ñ', 'Ïó∞Î†πÎ≥Ñ'], var_name='ÎèÑÏãú', value_name='Ïù∏ÏõêÏàò')

# Ïù∏ÏõêÏàòÎ•º Ïà´ÏûêÎ°ú Î≥ÄÌôò
pivot_df2_melted['Ïù∏ÏõêÏàò'] = pd.to_numeric(pivot_df2_melted['Ïù∏ÏõêÏàò'], errors='coerce')

# 'ÏÑ±Î≥Ñ'Ïù¥ÎùºÎäî ÏûòÎ™ªÎêú Í∞íÏùÑ ÌïÑÌÑ∞ÎßÅÌïòÏó¨ Ï†úÍ±∞
pivot_df2_melted = pivot_df2_melted[pivot_df2_melted['ÏÑ±Î≥Ñ'].isin(['ÎÇ®Ïûê', 'Ïó¨Ïûê'])]

# ÏÑ±Î≥Ñ Ïª¨ÎüºÏùò Ïã§Ï†ú Í∞íÏóê ÎßûÍ≤å palette ÏßÄÏ†ï
palette_dict = {'ÎÇ®Ïûê': 'blue', 'Ïó¨Ïûê': 'red'}

# ÏãúÍ∞ÅÌôî: ÎÇ®ÏûêÏôÄ Ïó¨ÏûêÎ≥Ñ Í≥†Î†πÏûê Ïàò Ìï©Ï≥êÏÑú ÌëúÏãú
plt.figure(figsize=(14, 8))
sns.barplot(x='ÎèÑÏãú', y='Ïù∏ÏõêÏàò', hue='ÏÑ±Î≥Ñ', data=pivot_df2_melted, palette=palette_dict, errorbar=None)
plt.title('ÎèÑÏãúÎ≥Ñ Í≥†Î†πÏûê Ïàò (ÏÑ±Î≥Ñ)', fontsize=16)
plt.xlabel('ÎèÑÏãú', fontsize=14)
plt.ylabel('Ïù∏ÏõêÏàò', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='ÏÑ±Î≥Ñ')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
