#📌 Q3
groupby를 사용하여 2020년의 온라인 해외판매 상품군을 지역별 합계를 구해 분석해 주세요.

2020년은 코로나로 인해 전국민이 어려움을 겪었던 시기입니다. 또, 아마존 등의 해외 온라인 판매가 늘어난 시기이기도 합니다. 이 때, 해외 어느지역에 온라인 판매가 얼마나 되었는지를 groupby() 와 unstack()을 활용해서 합계를 구하고 분석해 주세요! 또 분석한 결과를 보고 어떤 인사이트를 얻었는지도 함께 작성해 주세요!
그래프의 색상, 스타일은 자유롭게 원하는 스타일로 지정해도 됩니다.

```python
# 2020년 데이터만 필터링
df_2020 = df_filtered[(df_filtered['연도'] == 2020) ]

# '상품군별'과 '국가(대륙)별'로 그룹화하여 '백만' 열의 합계를 계산
grouped = df_2020.groupby(['국가(대륙)별', '상품군별'])['백만'].sum()

# 상위 7개 상품군별 선택 (전체 합계 기준)
top7_groups = grouped.groupby('상품군별').sum().nlargest(7).index

# 상위 7개 상품군만 필터링
grouped_top7 = grouped.loc[:, top7_groups]

# 결과를 지역별로 unstack 하여 넓은 형식으로 변환
grouped_unstacked = grouped_top7.unstack()

# 결과 시각화
grouped_unstacked.plot(kind='bar', figsize=(12, 6), colormap='tab10')

# 범례를 그래프 밖으로 빼기
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.title("2020년 지역별 온라인 해외판매 상위 7개 상품군별 합계")
plt.xlabel("국가(대륙)별")
plt.ylabel("판매액 (백만 원)")
plt.xticks(rotation=0)
plt.show()

# unstack, melt 비교
# unstack: 인덱스를 열로 변환합니다.
# 데이터를 넓은 형태로 변환합니다.
#주로 다중 인덱스를 처리할 때 사용됩니다.

# melt:열을 행으로 변환합니다.
# 데이터를 길고 좁은 형태로 변환합니다.
# 주로 데이터의 재구조화에 사용됩니다.

# 결론: unstack은 데이터를 넓게 변환하고, melt는 데이터를 길게 변환하는 데 사용됩니다.
# 이 두 함수는 서로 반대 방향의 작업을 수행하는 경우가 많으며,
# 데이터의 구조를 조정하거나 변형할 때 필요한 상황에 따라 적절히 사용됩니다.
```

```python
# 대륙별 판매액 합계 계산
continent_total = df_2020.groupby('국가(대륙)별')['백만'].sum()

# 원 그래프 시각화
plt.figure(figsize=(10, 8))
plt.pie(continent_total, labels=continent_total.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=140)

plt.title('2020년 대륙별 온라인 해외 판매액 비율', fontsize=16)
plt.show()
```

```python
# 지역별 상품군별 합계 계산 (누적)
pivot_data = df_2020.pivot_table(index='상품군별', columns='국가(대륙)별', values='백만', aggfunc='sum')

# 스택형 누적 면적 그래프 시각화
plt.figure(figsize=(14, 8))
pivot_data.plot(kind='area', stacked=True, alpha=0.6, colormap='tab20')

plt.title('2020년 상품군별 대륙별 온라인 판매액', fontsize=16)
plt.xlabel('상품군')
plt.ylabel('판매액 (백만)')
plt.legend(title='대륙')
plt.grid(True)
plt.show()
```

#📌 Q5
KOSIS 에 있는 데이터를 하나 정해 자유롭게 분석해 주세요!

지금까지 익힌 방법으로 다른 데이터에도 같은 방법을 통해 분석해 볼 수 있을 것입니다.
KOSIS(https://kosis.kr/) 사이트에서 관심있는 분야의 데이터를 찾고 원하는 시각화를 해주세요!
조회조건에 계, 합계 등을 미리 제거하고 다운로드 받으면 전처리를 줄일 수 있습니다.
행렬전환 기능을 사용하면 따로 melt를 진행하지 않아도 되기 때문에 이 또한 전처리를 줄일 수 있습니다.

```python
# 60세 이상 생활비 원천, 시점 2020년
file_path = '/성__연령_및_생활비_원천별_고령자_60세이상_시도_20240825200139.csv'

# CSV 파일을 읽어서 데이터프레임으로 변환합니다.
df2 = pd.read_csv(file_path, encoding='cp949')

# 데이터프레임의 상위 5개 행을 출력합니다.
df2.head()
```

```python
#도시별 고령화 수(성별) 분석

# 데이터프레임을 도시별로 pivot
pivot_df2 = df2.pivot_table(
    index=['성별', '연령별', '항목'],
    values=df2.columns[3:],  # 도시별 인원수 열 선택
    aggfunc='sum'  # 인원수 합계 계산
).reset_index()

# 시각화를 위한 준비
pivot_df2_melted = pivot_df2.melt(id_vars=['항목', '성별', '연령별'], var_name='도시', value_name='인원수')

# 인원수를 숫자로 변환
pivot_df2_melted['인원수'] = pd.to_numeric(pivot_df2_melted['인원수'], errors='coerce')

# '성별'이라는 잘못된 값을 필터링하여 제거
pivot_df2_melted = pivot_df2_melted[pivot_df2_melted['성별'].isin(['남자', '여자'])]

# 성별 컬럼의 실제 값에 맞게 palette 지정
palette_dict = {'남자': 'blue', '여자': 'red'}

# 시각화: 남자와 여자별 고령자 수 합쳐서 표시
plt.figure(figsize=(14, 8))
sns.barplot(x='도시', y='인원수', hue='성별', data=pivot_df2_melted, palette=palette_dict, errorbar=None)
plt.title('도시별 고령자 수 (성별)', fontsize=16)
plt.xlabel('도시', fontsize=14)
plt.ylabel('인원수', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='성별')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
